Churn by Contract Type

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=telco_data, x='Contract', hue='Churn')
plt.title('Churn by Contract Type')
plt.xticks(rotation=45)
plt.legend(title='Churn', labels=['No', 'Yes'])
plt.show()
```

Month-to-month contracts might have a higher churn rate, indicating a potential area for improvement in customer retention strategies for this contract type.
one-year contracts tend to have a higher likelihood of remaining with the company.
Two-year contracts also have a similar trend like the one year contract but better.

Churn by Additional Services

```python
services = ['OnlineSecurity', 'TechSupport', 'StreamingTV', 'StreamingMovies']
plt.figure(figsize=(12, 6))

for service in services:
    plt.subplot(2, 2, services.index(service) + 1)
    sns.countplot(data=telco_data, x=service, hue='Churn')
    plt.title(f'Churn by {service}')

plt.tight_layout()
plt.show()
```

Churn rate reduces with Online Security
With Tech Support the churn rate reduces

Churn by Monthly Charges and Total Charges

```python

plt.figure(figsize=(10, 6))

Mth = sns.kdeplot(telco_data["MonthlyCharges"][(telco_data["Churn"] == 0)],
                color="Red", shade=True)
Mth = sns.kdeplot(telco_data["MonthlyCharges"][(telco_data["Churn"] == 1)],
                ax=Mth, color="Blue", shade=True)

Mth.legend(["No Churn", "Churn"], loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('MonthlyCharges')
Mth.set_title('Monthly charges by churn')

plt.show()
```

From the above line graph, Churn is high when Monthly Charges are high
